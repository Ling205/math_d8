```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return x**3 - x

def bisection_method(func, a, b, tol, max_iter=100):
    data = {'Iteration': [], 'a': [], 'b': [], 'Length': []}
    iter_count = 0

    while (b - a) / 2 > tol and iter_count < max_iter:
        mid = (a + b) / 2
        f_mid = func(mid)

        data['Iteration'].append(iter_count)
        data['a'].append(a)
        data['b'].append(b)
        data['Length'].append(b - a)

        if f_mid == 0:
            return mid, pd.DataFrame(data), iter_count
        elif func(a) * f_mid < 0:
            b = mid
        else:
            a = mid

        iter_count += 1

    xmin = (a + b) / 2
    ymin = func(xmin)
    data['Iteration'].append(iter_count)
    data['a'].append(a)
    data['b'].append(b)
    data['Length'].append(b - a)
    df = pd.DataFrame(data)

    return xmin, df, iter_count

def plot_iterations_vs_tolerance(accuracies, method_func):
    iteration_counts = []
    for tol in accuracies:
        _, _, iterations = method_func(func, 0, 1, tol)
        iteration_counts.append(iterations)

    plt.plot(np.log10(accuracies), iteration_counts, marker='o')
    plt.title('Iterations vs Tolerance')
    plt.xlabel('log10(Tolerance)')
    plt.ylabel('Number of Iterations')
    plt.show()

# Пользователь вводит точность
tolerance = float(input("Введите точность: "))

# Применяем метод половинного деления
xmin, bisection_table, bisection_iterations = bisection_method(func, 0, 1, tolerance)

# Вывод результатов
print("\nМетод половинного деления:")
print("Минимум в виде f(xmin) =", xmin)
print("Количество итераций:", bisection_iterations)

# Вывод таблицы с результатами исследования метода половинного деления
print("\nТаблица результатов исследования метода половинного деления:")
print(bisection_table)

# Построение графика зависимости количества вычислений целевой функции от логарифма задаваемой точности
plot_iterations_vs_tolerance([10**(-i) for i in range(1, 10)], bisection_method)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return x**3 - x

def golden_ratio_method(func, a, b, tol, max_iter=100):
    data = {'Iteration': [], 'a': [], 'b': [], 'Length': []}
    iter_count = 0

    golden_ratio = (1 + 5 ** 0.5) / 2

    c = b - (b - a) / golden_ratio
    d = a + (b - a) / golden_ratio

    while (b - a) / 2 > tol and iter_count < max_iter:
        f_c = func(c)
        f_d = func(d)

        data['Iteration'].append(iter_count)
        data['a'].append(a)
        data['b'].append(b)
        data['Length'].append(b - a)

        if f_c < f_d:
            b = d
        else:
            a = c

        c = b - (b - a) / golden_ratio
        d = a + (b - a) / golden_ratio

        iter_count += 1

    xmin = (a + b) / 2
    ymin = func(xmin)
    data['Iteration'].append(iter_count)
    data['a'].append(a)
    data['b'].append(b)
    data['Length'].append(b - a)
    df = pd.DataFrame(data)

    return xmin, df, iter_count

# Пользователь вводит точность
tolerance = float(input("Введите точность: "))

# Применяем метод золотого сечения
xmin, golden_ratio_table, golden_ratio_iterations = golden_ratio_method(func, 0, 1, tolerance)

# Вывод результатов
print("\nМетод золотого сечения:")
print("Минимум в виде f(xmin) =", xmin)
print("Количество итераций:", golden_ratio_iterations)

# Вывод таблицы с результатами исследования метода золотого сечения
print("\nТаблица результатов исследования метода золотого сечения:")
print(golden_ratio_table)

# Построение графика зависимости количества вычислений целевой функции от логарифма задаваемой точности
plot_iterations_vs_tolerance([10**(-i) for i in range(1, 10)], golden_ratio_method)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return x**3 - x

def fibonacci_method(func, a, b, tol, max_iter=100):
    data = {'Iteration': [], 'a': [], 'b': [], 'Length': []}
    iter_count = 0

    def fibonacci(n):
        if n == 0:
            return 0
        elif n == 1:
            return 1
        else:
            a, b = 0, 1
            for _ in range(2, n + 1):
                a, b = b, a + b
            return b

    n = 1
    while fibonacci(n) <= (b - a) / tol:  
        n += 1

    rho = 1 - fibonacci(n - iter_count - 1) / fibonacci(max(1, n - iter_count))  
    c = a + rho * (b - a)
    d = a + (1 - rho) * (b - a)

    while iter_count < max_iter:
        f_c = func(c)
        f_d = func(d)

        data['Iteration'].append(iter_count)
        data['a'].append(a)
        data['b'].append(b)
        data['Length'].append(b - a)

        if f_c < f_d:
            b = d
        else:
            a = c

        rho = 1 - fibonacci(n - iter_count - 1) / fibonacci(max(1, n - iter_count))  
        c = a + rho * (b - a)
        d = a + (1 - rho) * (b - a)

        iter_count += 1

    xmin = (a + b) / 2
    ymin = func(xmin)
    data['Iteration'].append(iter_count)
    data['a'].append(a)
    data['b'].append(b)
    data['Length'].append(b - a)
    df = pd.DataFrame(data)

    return xmin, df, iter_count

# Пользователь вводит точность
tolerance = float(input("Введите точность: "))

# Применяем метод Фибоначчи
xmin, fibonacci_table, fibonacci_iterations = fibonacci_method(func, 0, 1, tolerance)

# Вывод результатов
print("\nМетод Фибоначчи:")
print("Минимум в виде f(xmin) =", xmin)
print("Количество итераций:", fibonacci_iterations)

# Вывод таблицы с результатами исследования метода Фибоначчи
print("\nТаблица результатов исследования метода Фибоначчи:")
print(fibonacci_table)

# Построение графика зависимости количества вычислений целевой функции от логарифма задаваемой точности
plot_iterations_vs_tolerance([10**(-i) for i in range(1, 10)], fibonacci_method)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def func(x):
    return x**3 - x

def derivative_func(x):
    return 3 * x**2 - 1

def newton_method(func, derivative_func, x0, tol=1e-6, max_iter=100):
    data = {'Iteration': [], 'x': [], 'f(x)': [], '|f(x)|': []}
    iter_count = 0
    x = x0

    while abs(func(x)) > tol and iter_count < max_iter:
        data['Iteration'].append(iter_count)
        data['x'].append(x)
        data['f(x)'].append(func(x))
        data['|f(x)|'].append(abs(func(x)))

        x = x - func(x) / derivative_func(x)

        iter_count += 1

    df = pd.DataFrame(data)

    return x, df, iter_count

# Пользователь вводит начальное приближение и точность
x0 = float(input("Введите начальное приближение: "))
tolerance = float(input("Введите точность: "))

# Применяем метод Ньютона
xmin, newton_table, newton_iterations = newton_method(func, derivative_func, x0, tolerance)

# Вывод результатов
print("\nМетод Ньютона:")
print("Минимум в виде f(xmin) =", xmin)
print("Количество итераций:", newton_iterations)

# Вывод таблицы с результатами исследования метода Ньютона
print("\nТаблица результатов исследования метода Ньютона:")
print(newton_table)

initial_guess = 1
accuracies = [10**(-i) for i in range(1, 10)]
iteration_counts = [newton_method(func, derivative_func, initial_guess, tol)[2] for tol in accuracies]

plt.plot(np.log10(accuracies), iteration_counts, marker='o')
plt.title('Iterations vs Tolerance')
plt.xlabel('log10(Tolerance)')
plt.ylabel('Number of Iterations')
plt.show()
```
